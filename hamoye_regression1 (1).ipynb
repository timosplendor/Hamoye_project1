# A linear  regression model to predict the temperatue of inner room base on the temperature of outside

# GETTING DATA

```python
import numpy as np
import pandas as pd
data=pd.read_csv( 'energydata_complete.csv') 
data.head(10)
```

# CHECKING FOR MISSING VALUE

```python
missing_values_count=data.isnull().sum()
missing_values_count
```

```python
data= data.drop(columns=['date','lights'])
```

```python
data
```

# SCALING THE DATA

```python
from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler() 
```

```python
norm_data = pd.DataFrame(scaler.fit_transform(data), columns=data.columns) 
```

```python
#Getting the independent and dependent variable
Temp_living_room =norm_data[['T2']]
Temp_outside =norm_data[['T6']]
```

```python
Temp_living_room
```

```python
Temp_outside
```

#  split and training the model

```python
from sklearn.model_selection import train_test_split 
T2_train,T2_test, T6_train,T6_test = train_test_split(Temp_living_room,Temp_outside,test_size= 0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model_T=LinearRegression()
model_T.fit(T2_train,T6_train)
```

```python
predicted_Temp_living_room = model_T.predict(T2_test)
predicted_Temp_living_room
```

# Evaluating the model

```python

from sklearn import metrics

meanAbErr = metrics.mean_absolute_error(T6_test,predicted_Temp_living_room)
meanSqErr = metrics.mean_squared_error(T6_test,predicted_Temp_living_room)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(T6_test,predicted_Temp_living_room))
r2=metrics.r2_score(T6_test, predicted_Temp_living_room)
rss = np.sum(np.square(T6_test-predicted_Temp_living_room)) 
```

```python
print("mean_absolute_error:",meanAbErr )
print("mean_squared_error:",meanSqErr)
print("root mean_squared_error:",rootMeanSqErr )
print("r2_score:",r2)
```
